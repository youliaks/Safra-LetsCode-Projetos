```python
import csv 
import re
import sys
import ast 
def obter_dados():
    '''
    Essa função carrega os dados dos musicos em um arquivo tipo csv
    '''
    nome_csv = 'teste_musicos_banda1.csv'
    with open(nome_csv, 'r', encoding="utf-8") as arquivo:
        dados = csv.reader(arquivo, delimiter=',', lineterminator='\n')  

    return dados 
```

```python
def validar_email(email: str):
    '''
    Função para validar os emails que o usuario digita
    '''
    regex = re.compile(r'([A-Za-z0-9]+[._])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')

    if re.fullmatch(regex, email):
        return(email)
    else:
        email = input("E-mail inválido. Por favor, insira outro e-mail: ")
```

```python
def cadastrar_musicos():
    '''
    Função para cadastrar os musicos no csv
    '''
    nome = input('Digite o nome do músico:').lower()
    email = input('Digite o email:').lower()
    csv_file = csv.reader(open('teste_musicos_banda1.csv', "r"), delimiter=",",lineterminator='\n')

    for row in csv_file:

        if email in row[0:]:
            print ('O e-mail já está cadastrado.')
            print (row)
            email = input(" Por favor, insira outro e-mail: ")
        else:
            continue
    email = validar_email(email)
    
    n_genero_musical = int(input('Digite a quantidade de gêneros musicais com que se identifica: '))
    if n_genero_musical < 1:
        n_genero_musical = int(input('Quantidade inválida! Digite a quantidade de gêneros musicais com que se identifica: '))
        lista_genero_musical = list(map(str,input("Digite o gênero musical (utilize espaço somente para separar os gêneros musicais): ").lower().strip().split()))[:n_genero_musical]
    else:
        lista_genero_musical = list(map(str,input("Digite o gênero musical (utilize espaço somente para separar os gêneros musicais): ").lower().strip().split()))[:n_genero_musical]
    
    n_instrumentos = int(input('Digite a quantidade de instrumentos que o músico toca: '))
    if n_instrumentos < 1:
        n_instrumentos = int(input('Quantidade inválida! Digite a quantidade de instrumentos que o músico toca: '))
        lista_instrumentos = list(map(str,input("Digite o o instrumento que toca (utilize espaço somente para separar os instrumentos): ").lower().strip().split()))[:n_instrumentos]
    else:
        lista_instrumentos = list(map(str,input("Digite o instrumento que toca (utilize espaço somente para separar os instrumentos): ").lower().strip().split()))[:n_instrumentos]
    

    return nome,email,lista_genero_musical,lista_instrumentos
```

```python
def append_lista(row_contents):
    '''
    Função para adicionar o musico que o usuariio cadastrou
    '''        
    with open('teste_musicos_banda1.csv', 'a+', newline='\n') as write_obj:
        csv_writer = csv.writer(write_obj)
        csv_writer.writerow(row_contents)    
    write_obj.close()  
```

```python
def buscar_musicos(opcao_busca):
    '''
    Função para buscar musicos de 3 maneiras
    '''
    csv_file = csv.reader(open('teste_musicos_banda1.csv', "r"), delimiter=",")    
    
    if opcao_busca == '1':
        opcao_1 = input('Digite o nome ou o e-mail ou o número correspondente ao gênero musical: ').lower()
        
        for row in csv_file:
            if opcao_1 in row[0] or opcao_1 in row[1] or opcao_1 in row[2]:
                print (f'Nome do músico: {row[0]} | E-mail: {row[1]} | Gênero Musical: {row[2]} | Instrumentos: {row[3]}')
            else:
                opcao_busca == '1'
    elif opcao_busca == '2':            
        opcao_2 = input('Digite o instrumento musical: ').lower()
        
        for row in csv_file:
            if opcao_2 in row[3]:
                print (f'Nome do músico: {row[0]} | E-mail: {row[1]} | Gênero Musical: {row[2]} | Instrumentos: {row[3]}')           

    elif opcao_busca == '3':
        opcao_3a = input('Digite o nome ou o e-mail ou o gênero musical: ').lower()
        opcao_3b = input('Digite o instrumento musical: ').lower()
        
        for row in csv_file:
            if opcao_3a in row[0] or opcao_3a in row[1] or opcao_3a in row[2] and opcao_3b in row[3]:
                print (f'Nome do músico: {row[0]} | E-mail: {row[1]} | Gênero Musical: {row[2]} | Instrumentos: {row[3]}')
```

```python
def filtrador(instrumento):
    '''
    Função para filtrar por instrumento uma lista já filtrada por genero
    '''
    def item_filtrador(lista_musicos):
        return lista_musicos[1] == instrumento
    return item_filtrador
```

```python
def gera_bandas(colecao, banda = [], combinacoes = []):
    '''
    Função para gerar as bandas possiveis de acordo com genero, numero de musicos e instrumentos
    '''
    instrumento = colecao.pop()
    for musico in instrumento:
        if musico in banda:
            continue

        banda.append(musico)
        if colecao:
            combinacoes = gera_bandas(colecao, banda, combinacoes)
        else:
            combinacoes.append(banda.copy())
        banda.pop()
    colecao.append(instrumento)
    return combinacoes
```

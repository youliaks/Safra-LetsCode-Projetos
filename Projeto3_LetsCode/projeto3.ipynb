## 2. Considerando a m√©dia de todas as vari√°veis, quais foram as 3 vari√°veis que mais mudaram ano a ano.(Sem contar o score ou ranking) üìÜ

Abaixo s√£o feitas as seguintes a√ß√µes: 

Atribui-se as colunas de interesse de cada ano a uma vari√°vel adequada e cria-se a coluna de ano;\
Concatena-se todas os df;\
Agrupa-se as linhas pelo ano;\
Calcula-se a m√©dia de cada vari√°vel ano a ano;\
Calcula-se as varia√ß√µes nominais e percentuais das vari√°veis de um ano ao ano seguinte;\
Apresenta-se a m√©dia, o desvio padr√£o e valores m√°ximo e m√≠nimo de cada vari√°vel em todos os anos amostrados (de 2015 a 2019)

```python
df_2015.info()
df_2015_numeros = df_2015.drop(columns = ['Country', 'Region', 'Happiness Rank', 'Happiness Score',
                                          'Standard Error', 'Dystopia Residual'])
df_2015_numeros['Year'] = '2015'
df_2015_numeros.head(2)
```

```python
df_2016.info()
df_2016_numeros = df_2016.drop(columns = ['Country', 'Region', 'Happiness Rank', 'Happiness Score',
                                          'Lower Confidence Interval', 'Upper Confidence Interval', 'Dystopia Residual'])
df_2016_numeros['Year'] = '2016'
df_2016_numeros.head(2)
```

```python
def trocar_colunas(df, coluna_1, coluna_2):
    colunas_lista = list(df.columns)
    x, y = colunas_lista.index(coluna_1), colunas_lista.index(coluna_2)
    colunas_lista[y], colunas_lista[x] = colunas_lista[x], colunas_lista[y]
    df = df[colunas_lista]

    return df
```

```python
df_2017.info()
df_2017_numeros = df_2017.drop(columns = ['Country', 'Happiness.Rank', 'Happiness.Score',
                                          'Whisker.high', 'Whisker.low', 'Dystopia.Residual'])
df_2017_numeros = trocar_colunas(df_2017_numeros, 'Generosity', 'Trust..Government.Corruption.')                                        
df_2017_numeros.columns = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom','Trust (Government Corruption)', 'Generosity']
df_2017_numeros['Year'] = '2017'
df_2017_numeros.head(2)
```

```python
df_2018.info()
df_2018_numeros = df_2018.drop(columns = ['Overall rank', 'Country or region', 'Score'])
df_2018_numeros = trocar_colunas(df_2018_numeros, 'Generosity', 'Perceptions of corruption')
df_2018_numeros.columns = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom','Trust (Government Corruption)', 'Generosity']
df_2018_numeros['Year'] = '2018'
df_2018_numeros.head(2)
```

```python
df_2019_numeros = df_2019.drop(columns = ['Overall rank', 'Country or region', 'Score'])
df_2019_numeros = trocar_colunas(df_2019_numeros, 'Generosity', 'Perceptions of corruption')
df_2019_numeros.columns = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom','Trust (Government Corruption)', 'Generosity']
df_2019_numeros['Year'] = '2019'
df_2019_numeros.head()
```

```python
df_5_anos = pd.concat([df_2015_numeros, df_2016_numeros, df_2017_numeros, df_2018_numeros, df_2019_numeros], axis = 0).reset_index(drop=True)
df_5_anos
```

```python
agrupado_por_ano = df_5_anos.groupby(by = 'Year')
agrupado_por_ano.mean().plot(ylim=(0,2.5), ylabel='Score per criterion').legend(loc='upper left')
media_variaveis_por_ano = agrupado_por_ano.mean()

display(media_variaveis_por_ano)
```

```python
variacao = media_variaveis_por_ano.diff().dropna().apply(lambda x: round(x,3))
variacao.index = ['De 2015 a 2016', 'De 2016 a 2017', 'De 2017 a 2018', 'De 2018 a 2019']
variacao
```

```python
variacao_percentual = (media_variaveis_por_ano.pct_change().dropna() * 100).apply(lambda x: round(x,2))
variacao_percentual.index = ['De 2015 a 2016 [%]', 'De 2016 a 2017 [%]', 'De 2017 a 2018 [%]', 'De 2018 a 2019 [%]']
variacao_percentual
```
